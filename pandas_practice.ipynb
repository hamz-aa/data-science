```python
# can change the index of rows
# i.e : normal index is 0, 1, 2, ...
# can change that to another column
movies_data = pd.read_csv('tmdb_5000_movies.csv', index_col = 'original_title')
```

```python
movies_data.head()
```

```python
# now loc can be used in a different way
movies_data.loc['Avatar']
```

```python
movies_data.loc[['Avatar', 'Spectre']]
```

```python
# can also access both rows and columns this way
movies_data.loc[['Avatar', 'Spectre'], ['budget', 'revenue']]
```
